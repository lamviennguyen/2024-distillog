```python
frames = [df1, df3]
data = pd.concat(frames)
```

```python
print(len(data))
```

```python
data = data.values
```

```python
import numpy as np

np.random.shuffle(data)
```

```python
split = int(len(data)*0.8)

print(split)
```

```python
train = []
test = []

for i in range(split):
  if data[i, 1] == 0:
    train.append(data[i,:])
  else:
    test.append(data[i,:])

for i in range(split, len(data)):
  test.append(data[i,:])
```

```python
print(len(train))
print(len(test))


print(len(train)+len(test))
```

```python
from collections import Counter

tmp = np.array(train)
label = tmp[:,1]
Counter(label)
```

```python
tmp = np.array(test)
label = tmp[:,1]
Counter(label)
```

```python
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)
train_df.to_csv('unsup_train.csv', header = None, index = None)
test_df.to_csv('unsup_test.csv', header = None, index = None)
```
